```python
from scipy.cluster.hierarchy import fcluster

def print_clusters(timeSeries, Z, k, plot=False):
    # k Number of clusters I'd like to extract
    results = fcluster(Z, k, criterion='maxclust')

    # check the results
    s = pd.Series(results)
    clusters = s.unique()

    for c in clusters:
        cluster_indeces = s[s==c].index
        print("Cluster %d number of entries %d" % (c, len(cluster_indeces)))
        if plot:
            plt.plot(timeSeries.T[:,cluster_indeces])
            plt.show()

print_clusters(series, links, 10, plot=True)
```

```python
series = []

f, axs = subplots(nrows=22, ncols=6, figsize=(16, 32))
for r in xrange(22):
    for c in xrange(6):
        i = r * 6 + c
        ax = axs[r, c]
        values = df[df.columns[i]].values
        min_index = np.min(np.argwhere(values > 0))
        values[min_index - 1:]
        max_first_6months = np.max(values[min_index:])
        column_series = values[min_index - 1:] / max_first_6months
#         ax.plot(df.index[min_index - 1:], column_series)
        ax.plot(column_series)
        series.append(column_series)
#         ax.xaxis.set_visible(False)
#         ax.yaxis.set_visible(False)
subplots_adjust()
```

```python
series = []

for col in df.columns:
    values = df[col].values
    min_index = np.min(np.argwhere(values > 0))
    values[min_index - 1:]
    max_first_6months = np.max(values[min_index:])
    column_series = values[min_index - 1:] / max_first_6months
    ts = df.index[min_index - 1:].values.astype(float)
    ts = (ts - ts[0]) / (24. * 60 * 60) / 1e9
    series.append((ts, column_series))
```

```python
series[0]
```

```python
ts_interp = np.arange(0, 10. * 5 * 30, 30)
series_interp = []
for ts, column_series in series:
    column_series_interp = np.interp(ts_interp, ts, column_series)
    series_interp.append(column_series_interp)
series_interp = np.array(series_interp)
```

```python
f, axs = subplots(nrows=22, ncols=6, figsize=(16, 32))
for r in xrange(22):
    for c in xrange(6):
        i = r * 6 + c
        ax = axs[r, c]
        values = series_interp[i]
        ax.plot(values)
#         ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)
subplots_adjust()
```
